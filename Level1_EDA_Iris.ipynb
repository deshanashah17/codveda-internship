```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")
```

```python
df = pd.read_csv("1) iris.csv")
df.head()
```

```python
print("Dataset Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())
```

```python

print("Missing Values:\n", df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
# Histogram
df.hist(figsize=(10, 6), bins=20, color='skyblue')
plt.suptitle("Feature Distributions")
plt.show()

# Pairplot (scatter plot matrix)
sns.pairplot(df, hue=df.columns[-1], palette="viridis")
plt.show()

# Boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Boxplots of Features")
plt.show()
```

```python
df.to_csv("iris_cleaned.csv", index=False)
```

```python
# Correlation matrix
corr = df.corr(numeric_only=True)
print(corr)

# Heatmap of correlations
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
Key Insights:
- Dataset has 150 samples, 3 species.
- No missing values found.
- Features show clear separation between species.
- Ready for further modeling.

Conclusion:
This EDA gives a clear overview of the Iris dataset for analysis and modeling.
```
